Aca solo se quiere acomodar los datos

```python
import os
import numpy as np
from skimage.io import imread, imshow, imsave
from skimage import img_as_ubyte
from skimage.transform import resize, rotate
from tqdm import tqdm
import json
```

```python
with open('configuration.json') as file:
    configuration = json.load(file)

TRAIN_PATH = f"{configuration["path"]["data_bowl"]}/{configuration["path"]["train"]}/"
TEST_PATH = f"{configuration["path"]["data_bowl"]}/{configuration["path"]["test"]}/"
SAVE_UNET = f"{configuration["path"]["data_y"]}/"
SAVE_AUTO = f"{configuration["path"]["data_z"]}/"
IMG_HEIGHT = configuration["image"]["heigth"]
IMG_WIDTH = configuration["image"]["width"]
IMG_CHANNELS = configuration["image"]["channels"]
```

```python
train_ids = next(os.walk(TRAIN_PATH))[1]
test_ids = next(os.walk(TEST_PATH))[1]

X_train = np.zeros((len(train_ids), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
Y_train = np.zeros((len(train_ids), IMG_HEIGHT, IMG_WIDTH, 1), dtype=np.bool)
lengths = []

for n, id in tqdm(enumerate(train_ids), total=len(train_ids)):
    path = TRAIN_PATH + id
    img = imread(path + "/images/" + id + ".png")[:,:,:IMG_CHANNELS]
    incorrect_size = img.shape[:2]!=(IMG_HEIGHT, IMG_WIDTH)
    if incorrect_size:
        img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
    X_train[n] = img
    mask = np.zeros((IMG_HEIGHT, IMG_WIDTH, 1), dtype=np.bool)
    for mask_file in next(os.walk(path + "/masks"))[2]:
        aux_mask = imread(path + "/masks/" + mask_file)
        if incorrect_size:
            aux_mask  = np.expand_dims(resize(aux_mask, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True), axis=-1)
        else:
            aux_mask = np.expand_dims(aux_mask, axis=-1)
        mask = np.maximum(mask, aux_mask)
    Y_train[n] = mask

X_test = np.zeros((len(test_ids), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.uint8)
for n, id in tqdm(enumerate(test_ids), total=len(test_ids)):
    path = TEST_PATH + id
    print(path + "/images/" + id + ".png")
    img = imread(path + "/images/" + id + ".png")[:,:,:IMG_CHANNELS]
    incorrect_size = img.shape[:2]!=(IMG_HEIGHT, IMG_WIDTH)
    if incorrect_size:
        img = resize(img, (IMG_HEIGHT, IMG_WIDTH), mode='constant', preserve_range=True)
    X_test[n] = img
```

```python
print(len(X_train))
print(len(X_test))
```

```python
length = len(X_test)
for n in tqdm(range(len(X_test)), total=len(X_test)):
    imsave(SAVE_AUTO+"image/"+str(n)+".png", X_test[n])
    imsave(SAVE_AUTO+"image/"+str(n+length)+".png", img_as_ubyte(rotate(X_test[n], 90)))
    imsave(SAVE_AUTO+"image/"+str(n+2*length)+".png", img_as_ubyte(rotate(X_test[n], 180)))
    imsave(SAVE_AUTO+"image/"+str(n+3*length)+".png", img_as_ubyte(rotate(X_test[n], 270)))

offset = length*4
length = len(X_train)
for n in tqdm(range(len(X_train)), total=len(X_train)):
    imsave(SAVE_AUTO+"image/"+str(offset+n)+".png", X_train[n])
    imsave(SAVE_AUTO+"image/"+str(offset+n+length)+".png", img_as_ubyte(rotate(X_train[n], 90)))
    imsave(SAVE_AUTO+"image/"+str(offset+n+2*length)+".png", img_as_ubyte(rotate(X_train[n], 180)))
    imsave(SAVE_AUTO+"image/"+str(offset+n+3*length)+".png", img_as_ubyte(rotate(X_train[n], 270)))
    imsave(SAVE_UNET+"image/"+str(n)+".png", X_train[n])
    imsave(SAVE_UNET+"image/"+str(n+length)+".png", img_as_ubyte(rotate(X_train[n], 90)))
    imsave(SAVE_UNET+"image/"+str(n+2*length)+".png", img_as_ubyte(rotate(X_train[n], 180)))
    imsave(SAVE_UNET+"image/"+str(n+3*length)+".png", img_as_ubyte(rotate(X_train[n], 270)))
    imsave(SAVE_UNET + "mask/" + str(n) + ".bmp", np.squeeze(Y_train[n]).astype(np.uint8) * 255)
    imsave(SAVE_UNET + "mask/" + str(n+length) + ".bmp", np.squeeze(img_as_ubyte(rotate(Y_train[n], 90)).astype(np.uint8) * 255))
    imsave(SAVE_UNET + "mask/" + str(n+2*length) + ".bmp", np.squeeze(img_as_ubyte(rotate(Y_train[n], 180)).astype(np.uint8) * 255))
    imsave(SAVE_UNET + "mask/" + str(n+3*length) + ".bmp", np.squeeze(img_as_ubyte(rotate(Y_train[n], 270)).astype(np.uint8) * 255))
```
